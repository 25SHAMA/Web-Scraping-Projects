## Scraping details from one page

```python
import requests
from bs4 import BeautifulSoup

link = 'https://books.toscrape.com/catalogue/page-1.html'
```

```python
res = requests.get(link)
```

```python
res
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup
```

```python
soup.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
len(soup.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3'))
```

```python
data = []

for book in soup.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3'):
   # print(book)

    book_link      =  'https://books.toscrape.com/catalogue/' + book.find_all('a')[-1].get('href')
    book_title     =  book.find_all('a')[-1].get('title')
    img_link       =  'https://books.toscrape.com/' + book.find('img').get('src')[3:]
    book_rating    =  book.find('p').get('class')[-1]
    price          =  book.find('p',class_ = 'price_color').text[1:]
    availability   =  book.find('p',class_ = 'instock availability').text.strip()
    
    data.append([book_title,book_rating,price,availability,book_link,img_link])
   
```

## Scraping the details from Multiple Pages

```python
import requests
from tqdm import tqdm
import pandas as pd
from bs4 import BeautifulSoup
```

```python
data = []

for i in tqdm(range(1,51)): #going through each page one by one
    link = 'https://books.toscrape.com/catalogue/page-' + str(i) + '.html' #creating link for each page

    res = requests.get(link)                              #sending requests to the link
    soup = BeautifulSoup(res.text, 'html.parser')         #creating a soup for that page

    for book in soup.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3'):
  

        book_link      =  'https://books.toscrape.com/catalogue/' + book.find_all('a')[-1].get('href')
        book_title     =  book.find_all('a')[-1].get('title')
        img_link       =  'https://books.toscrape.com/' + book.find('img').get('src')[3:]
        book_rating    =  book.find('p').get('class')[-1]
        price          =  book.find('p',class_ = 'price_color').text[1:]
        availability   =  book.find('p',class_ = 'instock availability').text.strip()
        
        data.append([book_title,book_rating,price,availability,book_link,img_link]) #saving scraped data in a list
    
    
```

```python
len(data)
```

```python
df = pd.DataFrame(data, columns = ['book_title','rating','price','availability','book_link','img_link'])
```
